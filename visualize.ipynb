## Downsample method comparison and visualization

```python
import matplotlib.pyplot as plt
import pandas as pd

ds1 = "output/train/larger_downsample/20221121-195355-convnext_xt_ds3x3-224/summary.csv"
ds2 = "output/train/20221128-105229-convnext_xt_ds2x2-224/summary.csv"
ds3 = "output/train/20221130-154310-convnext_xt_ds4x4g4-224/summary.csv"
ds4 = "output/train/20221128-160025-convnext_xt_dssep-224/summary.csv"

summary_paths = [ds1, ds2, ds3, ds4]
summary_titles = ["ds3x3", "ds2x2", "ds4x4g4", "dssep"]


def load_summary(file_path):
    df = pd.read_csv(file_path)
    return df[["eval_top1"]].to_numpy(), df[["eval_loss"]].to_numpy(), df[["train_loss"]].to_numpy()


train_summaries = []
for file_path in summary_paths:
    train_summaries.append(load_summary(file_path))


_, axes = plt.subplots(1, 2, figsize=(16, 5))

epochs = [0, min(len(summary[0]) for summary in train_summaries)]
x = list(range(*epochs))

axes[0].set_title("Top-1 Acc")
axes[1].set_title("Eval Loss")
for summary, title in zip(train_summaries, summary_titles):
    top1, eval_loss, _ = summary
    axes[0].plot(list(range(epochs[0], len(top1))), top1[epochs[0]:], label=title)
    axes[1].plot(list(range(epochs[0], len(eval_loss))), eval_loss[epochs[0]:], label=title)
axes[0].legend()
axes[1].legend()

print(epochs[1])
print(summary_titles)
print([max(top1[epochs[0]:epochs[1]]) for top1, _, _ in train_summaries])
print("eval loss:", [min(eval_loss[epochs[0]:epochs[1]]) for _, eval_loss, _ in train_summaries])
print("train loss:", [min(train_loss[epochs[0]:epochs[1]]) for _, _, train_loss in train_summaries])

plt.legend()
plt.show()
```

通过可视化原始的Downsample模块，我们可以看到，在使用稠密卷积做Downsample的时候，大部分参数都是0，即我们可以通过去除一些参数使得模型整体的计算量和参数量都有所下降而不损失性能。

最简单的方法是分组。分成4组，相当于参数量下降75%，此时性能不损失（甚至有提升）。这和剪枝方法的结果基本是接近的。

```python
import torch
import numpy as np
import matplotlib.pyplot as plt


ckpt_path = "output/train/20221128-105229-convnext_xt_ds2x2-224/last.pth.tar"
# ckpt_path = "output/train/larger_downsample/20221121-195355-convnext_xt-224/last.pth.tar"
ckpt = torch.load(ckpt_path)
# print(ckpt.keys())

model = ckpt["state_dict"]
ds_weight = model["downsample_layers.2.conv.weight"]
print(ds_weight.shape)

# _, axes = plt.subplots(10, 20, figsize=(16, 12))
# for r in range(10):
#     for c in range(20):
#         axes[r, c].imshow(ds_weight[r, c].cpu().numpy())
#         axes[r, c].axis("off")
# plt.show()

_, axes = plt.subplots(80, figsize=(16, 24))
for r in range(80):
    weight = ds_weight[r].permute(1, 2, 0).reshape(2, -1).cpu().numpy()
    axes[r].imshow(np.abs(weight))
    axes[r].axis("off")
    # print(np.abs(weight).min(), weight.mean(), weight.std())
plt.show()
```

## Kernel Decomposition

1. 当使用Hx1和1xW分解时，可以看到参数共用非常明显，即大部分卷积核都使用同一个Hx1和1xW卷积核；
2. 但是当Hx2和2xW分解时，可以看到二阶参数的差距非常大，很难说这里还有什么参数共用了，或者说很难和一阶同时出现参数共用。

### 结论

1. 可以使用参数共用，但目前影响不大。应优先考虑其他方面的优化
2. 水平和竖直分解可以很好地拟合局部卷积核。没必要使用“局部-全局串联”这种形式。即使要用也应该是“局部-全局并联”这种形式，因为串联形式会导致第二个卷积把第一个卷积做滑动平均，效果反而下降。

```python
import torch
import numpy as np
import matplotlib.pyplot as plt


ckpt_path = "checkpoint/convnext_tiny_22k_1k_224.pth"
ckpt = torch.load(ckpt_path)
# print(ckpt.keys())

model = ckpt["model"]
# print(model.keys())
ds_weight = model["stages.2.0.dwconv.weight"]
print(ds_weight.shape)


def svd(X):
    # Data matrix X, X doesn't need to be 0-centered
    n, m = X.shape
    # Compute full SVD
    U, Sigma, Vh = np.linalg.svd(X, full_matrices=False, compute_uv=True)
    ind = 2
    U = U[:, :ind]
    Sigma = Sigma[:ind]
    Vh = Vh[:ind, :]
    # Transform X with SVD components
    X_svd = np.dot(U, np.diag(Sigma))
    X_svd = np.dot(X_svd, Vh)
    print(np.mean((X_svd - X) ** 2) / np.mean(X ** 2))
    return X_svd, U, Vh


_, axes = plt.subplots(8, 12, figsize=(16, 12))
for i in range(8):
    for j in range(12):
        weight = ds_weight[i*12 + j, 0].cpu().numpy()
        w, U, Vh = svd(weight)
        # weight = np.abs(weight)
        axes[i, j].imshow(Vh, vmin=-.8, vmax=.8)
        axes[i, j].axis("off")
        # print(np.abs(weight).min(), weight.mean(), weight.std())
plt.show()
```
